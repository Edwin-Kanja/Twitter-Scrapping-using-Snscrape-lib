```python
import snscrape.modules.twitter as sntwitter
import pandas as pd



# Creating list to append tweet data to
tweets_list2 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('#Datascience since:2022-01-01 until:2022-06-25').get_items()):

    if i>500:
        break
    tweets_list2.append([tweet.date, tweet.id, tweet.content, tweet.url,\
                         tweet.user.username, tweet.user.followersCount,tweet.replyCount,\
                        tweet.retweetCount, tweet.likeCount, tweet.quoteCount, tweet.lang,\
                        tweet.outlinks, tweet.media, tweet.retweetedTweet, tweet.quotedTweet,\
                        tweet.inReplyToTweetId, tweet.inReplyToUser, tweet.mentionedUsers,\
                         tweet.coordinates, tweet.place, tweet.hashtags, tweet.cashtags])
# Creating a datafram from the tweets list above
tweets_df2 = pd.DataFrame(tweets_list2, columns=['Datetime', 'Tweet Id', 'Text', 'link', 'Username', 'followers', 'replies', 'retweets',
                                                'likes', 'qoutes', 'Lang', 'outlink', 'media', 'Retweeted tweets', 'Quoted tweet', 'inReplyToTweetId',
                                                'inReplyToUser', 'mentionedUsers', 'coordinates', 'place', 'hashtags', 'cashtags'])
tweets_df2
```

```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

# Creating list to append tweet data to
tweets_list1 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('from:@SafaricomPlc since:2022-04-30 until:2022-06-01').get_items()):
    if i>5000:
        break
    tweets_list1.append([tweet.date, tweet.id, tweet.content, tweet.url,\
                         tweet.user.username, tweet.user.followersCount, tweet.replyCount,\
                        tweet.retweetCount, tweet.likeCount, tweet.quoteCount, tweet.lang,\
                        tweet.outlinks, tweet.media, tweet.retweetedTweet, tweet.quotedTweet,\
                        tweet.inReplyToTweetId, tweet.inReplyToUser, tweet.mentionedUsers,\
                         tweet.coordinates, tweet.place, tweet.hashtags, tweet.cashtags])
    
# Creating a dataframe from the tweets list above 
tweets_df1 = pd.DataFrame(tweets_list1, columns=['Datetime', 'Tweet Id', 'Text', 'link', 'Username', 'followers', 'replies', 'retweets',
                                                'likes', 'qoutes', 'Lang', 'outlink', 'media', 'Retweeted tweets', 'Quoted tweet', 'inReplyToTweetId',
                                                'inReplyToUser', 'mentionedUsers', 'coordinates', 'place', 'hashtags', 'cashtags'])
tweets_df1
```

```python
#SEPARATE DATE AND TIME
import datetime as dt
tweets_df1['Date']=pd.to_datetime(tweets_df1['Datetime']).dt.date
tweets_df1['time']=pd.to_datetime(tweets_df1['Datetime']).dt.time
tweets_df1
```
